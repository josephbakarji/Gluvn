```python
def generate_num2note_dict():
    midi_numbers = list(range(24, 128))  # MIDI numbers range from 0 to 127
    octaves = list(range(0, 10))
    notes_sharp = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
    # Generate the note names for all octaves and associate them with their corresponding MIDI numbers

    Note2Midi = {}
    Num2Midi = {}
    for octave in octaves:
        for note, midi in zip(notes_sharp, midi_numbers[octave*12:(octave+1)*12]):
            Note2Midi[f"{note}{octave}"] = midi
            Num2Midi[midi] = f"{note}{octave}"

    return Note2Midi

generate_num2note_dict()
```
